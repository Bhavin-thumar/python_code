```python
import os, time
import pandas as pd
```

```python
os.chdir(r"F:\FEB-2020")
```

```python
base_path = "F:\FEB-2020"
```

```python
amt = []
name = []
tim = []
for file in os.listdir():
    a,*b,c = file.split('-')
    amt.append(a)
    name.append(c[:-4])
    t = time.ctime(os.path.getctime(file))
    t = t.split()[1:3]
    t = f'{t[0]}-{t[1]}'
    tim.append(t)
```

```python
df = pd.DataFrame({'amount' : amt, 
                   'particulars' : name,
                  'time' : tim})
df = df.astype({'amount' : int})
```

```python
df.groupby(by=['particulars','time'])['amount'].sum()
```

```python
df.sort_values(by= 'time')
```

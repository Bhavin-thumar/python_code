```python
from collections import deque
```

```python
class Stack:
    def __init__(self):
        self.container = deque()
        self.size = 0
        
    def push(self, data):
        self.size += 1
        return self.container.append(data)
    
    
    def pop(self):
        self.size -= 1
        return self.container.pop()
    
    def peek(self):
        return self.container[self.size-1]
    
    def __len__(self):
        return len(self.container)
    
    def is_empty(self):
        return len(self.container) == 0
    
    def reversee(self):
        return self.container.reverse()
```

```python
def infix_to_postfix(expression):
    operator = '+-*/^()'
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3, '(' : 4}
    output_string = ''
    s = Stack()
    for c in expression:
        print(f'Now : {c}')
        print(f'stack was : {s.container}, output was : {output_string}')
        if c not in operator:
            output_string += c
        else:
            if s.is_empty():
                s.push(c)
            else:
                if c != ')':
                    if precedence[c] > precedence[s.peek()] or s.peek() == '(':
                        s.push(c)
                    else:
                        while precedence[c] <= precedence[s.peek()]:
                            output_string += s.pop()
                            if s.is_empty() or s.peek() == '(':
                                s.push(c)
                                break
                else:
                    while s.peek() != '(':
                        output_string += s.pop()
                    s.pop()
        print(f'stack is : {s.container}, output is : {output_string}')
        print('----------------------------------------------------')
    while not s.is_empty():
        output_string += s.pop()
        
        
    return output_string
                                
```

```python
expression = 'a+b*(c^d-e)^(f+g*h)-i'
infix_to_postfix(expression)
```

```python
import cv2
import numpy as np
import os
```

```python
test_image = [os.path.join(path,jpg) 
             for path,folder,files in os.walk(r"C:\Users\hardip thummar\Desktop\face\collection")
             if files is not None
             for jpg in files]
#              if jpg.endswith('jpg')]
```

```python
label_dict = {}
for i in os.listdir(r"C:\Users\hardip thummar\Desktop\face\training_data"):
    k, v = i.split('-')
    label_dict[int(k)] = v
```

```python
def get_faces_and_labels(folder_path):
    face_list = []
    labels = []
    face_cascade = cv2.CascadeClassifier(r"C:\Users\hardip thummar\AppData\Local\Programs\Python\Python38-32\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
    for label in os.listdir(folder_path):
        folder = os.path.join(folder_path, label)
        for files in os.listdir(folder):
            img_path = os.path.join(folder, files)
            img = cv2.imread(img_path)
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 5)
            (x,y,w,h) = faces[0]
            cropped_face = gray[y:y+w, x:x+h]
            face_list.append(cropped_face)
            labels.append(int(label.split('-')[0]))
    return face_list, labels
```

```python
face_list, labels = get_faces_and_labels(r"C:\Users\hardip thummar\Desktop\face\training_data")
```

```python
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
face_recognizer.train(face_list, np.array(labels))
```

```python
face_cascade = cv2.CascadeClassifier(r"C:\Users\hardip thummar\AppData\Local\Programs\Python\Python38-32\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
```

```python
for images in test_image:
    face_in_image = []
    img = cv2.imread(images)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.15, 5)
    for x,y,w,h in faces:
        roi = gray[y:y+w, x:x+h]
        face_in_image.append(roi)
    print(f'No. of faces found in {images} - {len(faces)}.')
    for index,f in enumerate(face_in_image):
        id_, confidence = face_recognizer.predict(f)
        print(f'No. {index+1} face have : id - {label_dict[id_]}, confidence - {confidence}')
    print('--------------------------------------------------------------------')
```

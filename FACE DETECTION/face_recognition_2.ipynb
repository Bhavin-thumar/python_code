```python
bulk_df = concat_df[concat_df['Order ID'].duplicated(keep = False)]
bulk_df.reset_index(drop = True, inplace = True)
```

```python
bulk_df.head()
```

```python
product_group = dict()
sub_group = []
for row in bulk_df.itertuples():
    index, order_id, product = row[0], row[1], row[2]
    if index == 0:
        sub_group.append(product)
    else:
        if order_id == bulk_df.at[index-1,'Order ID']:
            sub_group.append(product)
        else:
            product_group[bulk_df.at[index-1, 'Order ID']] = sub_group
            sub_group = []
            sub_group.append(product)
    if index == len(bulk_df)-1:
        product_group[bulk_df.at[index, 'Order ID']] = sub_group
```

```python
from itertools import combinations
```

```python
pair = []
for group in product_group.values():
    pairs = combinations(group, 2)
    for item in pairs:
        pair.append(item)
```

```python
from collections import Counter
```

```python
c = Counter(pair)
```
